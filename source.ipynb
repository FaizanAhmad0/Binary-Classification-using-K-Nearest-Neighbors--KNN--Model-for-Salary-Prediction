```python
import pandas as pd
import numpy as np
import math   
from statistics import mode
```

```python
df=pd.read_excel("Data.xlsx")
df2=pd.read_excel("test.xlsx")
```

```python
dummy_df = pd.get_dummies(df)
dummy_df2 = pd.get_dummies(df2)
dummy_df2
```

```python
dummy_df["Dif"]=dummy_df["Salary"]
dummy_df.loc[(dummy_df.Dif < 35000), 'Dif'] = 0
dummy_df.loc[(dummy_df.Dif >= 35000), 'Dif'] = 1
```

```python
dummy_df.drop('Salary', inplace=True, axis=1)
dummy_df2.drop('Salary', inplace=True, axis=1)
```

```python
def returner(n):
    distance = []
    j=dummy_df2.iloc[n,0:24]
    for k in range(0,1000):
        l=dummy_df.iloc[k,0:24]
        distance.append(math.dist(j, l))
        
    temp = distance.copy()
    temp.sort()
    checker = []
    for i in range(10):
        ind = distance.index(temp[i])
        checker.append(dummy_df.iloc[ind]['Dif'])
    a = mode(checker)
    return a
```

```python
final = []
for i in range(0,20):
    final.append(returner(i))
dummy_df2["Salary"]=final
```
